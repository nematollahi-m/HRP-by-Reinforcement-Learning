```python
#!/usr/bin/env python3.7

# Copyright 2022, Gurobi Optimization, LLC

import gurobipy as gp
from gurobipy import GRB

try:

    # Create a new model
    m = gp.Model("mip1")

    # Create variables
    x1 = m.addVar(vtype=GRB.INTEGER, name="x1")
    x2 = m.addVar(vtype=GRB.INTEGER, name="x2")
    x3 = m.addVar(vtype=GRB.INTEGER, name="x3")
    
    z1 = m.addVar(vtype=GRB.INTEGER, name="z1")
    z2 = m.addVar(vtype=GRB.INTEGER, name="z2")
    z3 = m.addVar(vtype=GRB.INTEGER, name="z3")

    # Set objective
    m.setObjective(100*(x1+x2+x3)+35*(z1+z2+z3), GRB.MINIMIZE)

    # Add constraint: 
    m.addConstr(z1==x1, "z1")
    
    # Add constraint: 
    m.addConstr(z2==x2+z1, "z2")
    
     # Add constraint: 
    m.addConstr(z3==x3+z2, "z3")


     # Add constraint: 
    m.addConstr(z1<=10, "x1")
    
     # Add constraint: 
    m.addConstr(z2<=10, "x2")
    
     # Add constraint: 
    m.addConstr(z3<=10, "x3")
    
     # Add constraint: 
    m.addConstr(100*(x1+x2+x3)+35*(z1+z2+z3)<=5000, "b")
    
     # Add constraint: 
    m.addConstr(700*(z1+z2+z3)>=10000, "p")

    # Optimize model
    m.optimize()

    for v in m.getVars():
        print('%s %g' % (v.VarName, v.X))

    print('Obj: %g' % m.ObjVal)

except gp.GurobiError as e:
    print('Error code ' + str(e.errno) + ': ' + str(e))

except AttributeError:
    print('Encountered an attribute error')
```
